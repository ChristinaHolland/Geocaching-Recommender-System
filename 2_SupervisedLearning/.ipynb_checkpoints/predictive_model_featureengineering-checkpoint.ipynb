```python
def run_model(train, test, label):
    rf = RandomForestClassifier()
    rf.fit(train,y_train)
    preds = rf.predict(test)
    cm = confusion_matrix(y_test,preds)
    acc = (cm[0][0]+cm[1][1])/(cm[0][0]+cm[0][1]+cm[1][0]+cm[1][1])
    rec = cm[1][1]/(cm[1][1] + cm[1][0])
    prc = cm[1][1]/(cm[1][1] + cm[0][1])
    f1  = 2*prc*rec/(prc+rec)
    print(label + f': Accuracy={acc}, f1-Score={f1}, Recall={rec}, Precision={prc}')
    print('')
```

```python
# With default params:

run_model(X_1__train, X_1__test, 'original only')
run_model(X__2_train, X__2_test, 'squared only')
run_model(X___3train, X___3test, 'cross terms only')
run_model(X_12_train, X_12_test, 'original + squared')
run_model(X_1_3train, X_1_3test, 'original + cross')
run_model(X__23train, X__23test, 'squared + cross')
run_model(X_123train, X_123test, 'original, squared, + cross')
```

Using only the 2nd degree terms (squared and cross-terms) seems to do pretty well for the default parameters.

```python
X_train = X__23train
X_test  = X__23test

n_est_list = [10, 50, 100, 500, 1000, 5000]
max_depth_list = [2, 3, 4, 5, 6, 10, 20, None]


scores = []

for n_est in n_est_list:
    for max_depth in max_depth_list:
        
        modellabel = f'RandomForest n_est={n_est}, max_depth={max_depth}'

        rf = RandomForestClassifier(n_estimators=n_est, max_depth=max_depth)
        
        rf.fit(X_train,y_train)
        pred = rf.predict(X_test)
        cm = confusion_matrix(y_test,pred)
        acc = (cm[0][0]+cm[1][1])/(cm[0][0]+cm[0][1]+cm[1][0]+cm[1][1])
        rec = cm[1][1]/(cm[1][1] + cm[1][0])
        prc = cm[1][1]/(cm[1][1] + cm[0][1])
        f1  = 2*prc*rec/(prc+rec)
        scores.append({
                'model'        : modellabel,
                'accuracy'     : acc,
                'recall'       : rec,
                'precision'    : prc,
                'f1-score'     : f1
            })
        print(modellabel)
        print(f'ACC: {acc}, REC: {rec}, PRC: {prc}, F1S: {f1}')
```

```python
rf_scores = pd.DataFrame(scores)
rf_scores.sort_values(by='f1-score', ascending=False).head()
```
