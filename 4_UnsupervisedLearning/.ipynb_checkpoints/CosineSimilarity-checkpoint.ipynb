## Cosine Similarity Analysis

```python

from sklearn.preprocessing import StandardScaler
from sklearn.feature_extraction.text import CountVectorizer

import numpy as np
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt
from scipy import sparse
from sklearn.metrics.pairwise import pairwise_distances, cosine_distances, cosine_similarity
```

```python
data = pd.read_csv('../data/working_files/strippeddata.csv')
data.head()
```

```python
K0 = '0918273645.1425367.908.9012783465'

KEY = [pd.to_numeric(K0[7] +K0[0] +K0[24] +K0[20]+K0[10]+K0[8]), 
       pd.to_numeric(K0[13]+K0[20]+K0[0]  +K0[24]+K0[10]+K0[13]), 
       pd.to_numeric(K0[23]+K0[24]+K0[20] +K0[0] +K0[22]+K0[29])]

def OtherInfotoCode(z_list):
    A = [np.power(zz,0.5) for zz in z_list]
    c_list = [np.round(KEY[2]+aa) if aa%1>0.5 else np.round(KEY[2]-aa) for aa in A]
    c_list = [int(cc) for cc in c_list]
    return c_list

def NametoCode(n_list):
    A = [np.power(nn,0.5) for nn in n_list]
    c_list = [np.round(KEY[0]+aa) if aa%1>0.5 else np.round(KEY[0]-aa) for aa in A]
    c_list = [int(cc) for cc in c_list]
    return c_list

Zindex = list(data['index'])
Cindex = OtherInfotoCode(Zindex)
data['index'] = Cindex
data = data.sort_values(by='index')
data.set_index('index',inplace=True)
data.head()
```

```python
CT = pd.read_csv('../data/working_files/clean_text.csv')
Nindex = list(CT['name_index'])
Cindex = NametoCode(Nindex)
CT['index'] = Cindex
CT = CT.sort_values(by='index')
CT.reset_index(drop=True, inplace=True)
CT = CT[['index', 'clean_text']]
CT.head()
```

#### Need to reform data so that I have a col for cache id, a col for parameter id, and a col for (scaled) val. Will be 696248 x 3

```python
ndx = list(range(data.shape[0]))

cols = data.columns
ndx_long = []
param_long = []
val_long = []
for n,col in enumerate(cols):
    
    ndx_long   += ndx
    
    vals = list(data[col])
    mn = np.mean(vals)
    sd = np.std(vals)
    vals = [(v-mn)/sd for v in vals]
    val_long   += vals
    
    param_long += [n for p in vals]

data_new = pd.DataFrame({'ndx': ndx_long, 'param': param_long, 'vals': val_long})
data_new.shape
```

```python
pivot1 = data_new.pivot_table(index='ndx', columns = 'param', values = 'vals')
pivot1.shape
```

```python
pivot1.head()
```

```python
sparse_pivot1 = sparse.csr_matrix(pivot1.fillna(0))
```

```python
recommender1 = pairwise_distances(sparse_pivot1, metric='cosine')
recommender1_df = pd.DataFrame(recommender1, index=pivot1.index, columns=pivot1.index)
recommender1_df.shape
```

```python
recommender1_df.head()
```

```python
recommender1_df.min().min(), recommender1_df.max().max()
```

```python
ndx = list(range(data.shape[0]))

cols = words.columns
ndx_long = []
param_long = []
val_long = []
for n,col in enumerate(cols):
    
    ndx_long   += ndx
    
    vals = list(words[col])
    mn = np.mean(vals)
    sd = np.std(vals)
    vals = [(v-mn)/sd for v in vals]
    val_long   += vals
    
    param_long += [n for p in vals]

    if n%20==0: print(f'Appending column #{n}, ' + col)
        
print('Finished appending.')
```

Note:
    
cosine distance = 1 - cosine similarity. 

So,

| cos simularity | cos distance |
| --- | --- |
| -1 | 2 |
|  0 | 1 | 
|  1 | 0 |

and so my values of 0-1.8 or so make sense.

```python
# compare the two recommenders:

correlations = []
for n in range(12433):
    CD1 = list(recommender1_df[n])
    CD2 = list(recommender2_df[n])
    chk = pd.DataFrame({'by params': CD1, 'by text': CD2})
    correlations.append(chk.corr()['by params'][1])
    if n%500==0: print(n)
```

#### The NULL hypothesis would be that these two recommenders are completely uncorrelated, in which case the correlations should be centered on 0. Let's test that:

```python
bins = np.linspace(-1,1,51)
plt.hist(correlations,bins=bins);
plt.grid()
```

```python
sum([1 for c in correlations if c>0])*100/len(correlations)
```

88% of the caches have a correlation between the two recommenders of >0.

```python
# confidence interval:
```

```python
sample_mean = np.mean(correlations)
sigma = np.std(correlations)
n = len(correlations)
m_error = 1.96*sigma/np.sqrt(n)
ul = sample_mean + m_error
ll = sample_mean - m_error
print([ll, ul])
print(f'{sample_mean} +/- {m_error}')
```

To 95% confidence, the correlation between these two recommender dataframes is between 0.1052 and 0.1090. 

So, there is some connection, but they do each bring something to the table, it looks like.

```python
# Let's try combining into a single recommender:


ndx = list(data_new['ndx']) + list(words_new['ndx'])

params1 = list(data_new['param'])
params2 = list(words_new['param'])
params2 = [p+56 for p in params2]
params = params1 + params2

vals1 = list(data_new['vals'])
vals2 = list(words_new['vals'])
vals2 = [v/2 for v in vals2] # To somewhat reduce the influence of the text vs the parameters
vals = vals1 + vals2


matrix_new = pd.DataFrame({'ndx': ndx, 'param': params, 'vals': vals})
matrix_new.shape
```

```python
pivot_comb = matrix_new.pivot_table(index='ndx', columns = 'param', values = 'vals')
pivot_comb.shape
```

```python
pivot_comb.head()
```

```python
sparse_pivot = sparse.csr_matrix(pivot_comb.fillna(0))
```

```python
recommender_comb = pairwise_distances(sparse_pivot, metric='cosine')
recommender_comb_df = pd.DataFrame(recommender_comb, index=pivot_comb.index, columns=pivot_comb.index)
recommender_comb_df.shape
```

```python
recommender_comb_df.head()
```

```python
# compare the three recommenders:

correlations12 = []
correlations1_ = []
correlations_2 = []

for n in range(12433):
    CD1 = list(recommender1_df[n])
    CD2 = list(recommender2_df[n])
    CD3 = list(recommender_comb_df[n])
    chk = pd.DataFrame({'by params': CD1, 'by text': CD2, 'combined': CD3})
    correlations12.append(chk.corr()['by params'][1])
    correlations1_.append(chk.corr()['by params'][2])
    correlations_2.append(chk.corr()['by text'][2])
    if n%500==0: print(n)
```

```python
bins = np.linspace(-1,1,51)

fig = plt.figure(figsize=(12,4));
ax = fig.add_subplot(131);
plt.hist(correlations12,bins=bins);
plt.title('params vs text')
plt.grid()
ax = fig.add_subplot(132);
plt.hist(correlations1_,bins=bins);
plt.title('params vs combined')
plt.grid()
ax = fig.add_subplot(133);
plt.hist(correlations_2,bins=bins);
plt.title('text vs combined')
plt.grid()
```

```python
sample_mean = np.mean(correlations12)
sigma = np.std(correlations12)
n = len(correlations12)
m_error = 1.96*sigma/np.sqrt(n)
ul = sample_mean + m_error
ll = sample_mean - m_error
print('params vs text')
print([ll, ul])
print(f'{sample_mean} +/- {m_error}')
print()

sample_mean = np.mean(correlations1_)
sigma = np.std(correlations1_)
n = len(correlations1_)
m_error = 1.96*sigma/np.sqrt(n)
ul = sample_mean + m_error
ll = sample_mean - m_error
print('params vs combined')
print([ll, ul])
print(f'{sample_mean} +/- {m_error}')
print()

sample_mean = np.mean(correlations_2)
sigma = np.std(correlations_2)
n = len(correlations_2)
m_error = 1.96*sigma/np.sqrt(n)
ul = sample_mean + m_error
ll = sample_mean - m_error
print('text vs combined')
print([ll, ul])
print(f'{sample_mean} +/- {m_error}')
print()
```

To 95% confidence, the combined recommender is correlated to the parameter only recommender to between 0.880 and 0.884, and to the text only recommender to between 0.451 and 0.456.

Let's look at some sample results, and compare:

```python
L = 100

overlap12 = []
overlap1_ = []
overlap_2 = []
s_overlap12 = []
s_overlap1_ = []
s_overlap_2 = []

for n in range(12433):
    
    CD1 = recommender1_df[[n]]
    CD1 = CD1.sort_values(by=n)
    list1 = CD1.index[1:(L+1)]
    slist1 = list1[0:10]
    CD2 = recommender2_df[[n]]
    CD2 = CD2.sort_values(by=n)
    list2 = CD2.index[1:(L+1)]
    slist2 = list2[0:10]
    CD3 = recommender_comb_df[[n]]
    CD3 = CD3.sort_values(by=n)
    list3 = CD3.index[1:(L+1)]
    slist3 = list3[0:10]
    
    overlap12.append(sum([1 for c in list1 if c in list2]))
    overlap1_.append(sum([1 for c in list1 if c in list3]))
    overlap_2.append(sum([1 for c in list2 if c in list3]))
    s_overlap12.append(sum([1 for c in slist1 if c in slist2]))
    s_overlap1_.append(sum([1 for c in slist1 if c in slist3]))
    s_overlap_2.append(sum([1 for c in slist2 if c in slist3]))

    if n%500==0: print(n)
```

```python
chk = pd.DataFrame({
    'overlap 100 param v text': overlap12,
    'overlap 100 param v combo': overlap1_,
    'overlap 100 combo v text': overlap_2,
    'overlap 10 param v text': s_overlap12,
    'overlap 10 param v combo': s_overlap1_,
    'overlap 10 combo v text': s_overlap_2
                   })

chk.describe()
```
