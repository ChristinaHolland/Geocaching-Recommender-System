```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import scipy.stats as stats


alpha_prior = 1
beta_prior  = 1
prior_distn = stats.beta(alpha_prior, beta_prior)

yes_new = 6
no_new = 1
num = yes_new + no_new
alpha_posterior = yes_new + alpha_prior
beta_posterior = no_new + beta_prior
posterior_distn1 = stats.beta(alpha_posterior, beta_posterior)


mode_beta = np.round((alpha_posterior - 1)/(alpha_posterior + beta_posterior - 2),4)

pct_list = np.linspace(0,1,1001)

fig = plt.figure(figsize=(12,5));
plt.plot(pct_list, prior_distn.pdf(pct_list), label='prior (before survey)');
plt.plot(pct_list, posterior_distn1.pdf(pct_list), label=f'posterior (after {num} responses)');
plt.plot([mode_beta, mode_beta],[0, 3])
plt.title('Beta distribution of average favorability (1 = would use app, 0 = would not)')
plt.legend();
plt.grid();

plt.xlabel(f'Posterior Mode = (a-1)/(a+b-2) = {mode_beta}');
```
