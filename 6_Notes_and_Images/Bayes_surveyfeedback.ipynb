```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import scipy.stats as stats


alpha_prior = 1
beta_prior  = 1
prior_distn = stats.beta(alpha_prior, beta_prior)

yes_new = 2
no_new = 1
num = yes_new + no_new
alpha_posterior = yes_new + alpha_prior
beta_posterior = no_new + beta_prior
posterior_distn1 = stats.beta(alpha_posterior, beta_posterior)
mode_beta1 = np.round((alpha_posterior - 1)/(alpha_posterior + beta_posterior - 2),4)

yes_new = 5
no_new = 0
num = yes_new + no_new
alpha_posterior = yes_new + alpha_prior
beta_posterior = no_new + beta_prior
posterior_distn2 = stats.beta(alpha_posterior, beta_posterior)
mode_beta2 = np.round((alpha_posterior - 1)/(alpha_posterior + beta_posterior - 2),4)

pct_list = np.linspace(0,1,1001)

fig = plt.figure(figsize=(12,5));

plt.plot(pct_list, prior_distn.pdf(pct_list), color='black', label='prior (before survey or post)');
plt.plot(pct_list, posterior_distn1.pdf(pct_list), color='blue', label='3 survey responses');
plt.plot(pct_list, posterior_distn2.pdf(pct_list), color='orange', label='5 reddit upvotes');
plt.plot([mode_beta1, mode_beta1],[2, 3.5], color='blue')
plt.plot([mode_beta1-0.05, mode_beta1, mode_beta1+0.05],[2.5, 2, 2.5], color='blue')
plt.plot([mode_beta2, mode_beta2],[3.8, 5.2], color='orange')
plt.plot([mode_beta2-0.05, mode_beta2, mode_beta2+0.05],[4.7, 5.2, 4.7], color='orange')
plt.title('Beta distribution of average favorability (1 = would use app, 0 = would not)')
plt.xlabel(f'Posterior MODE = (a-1)/(a+b-2) = {mode_beta1} or {mode_beta2}, depending on survey vs post');
plt.legend();
plt.grid();

```
